```python

import pandas as pd

df = pd.read_csv(r"C:\Users\DELL\Downloads\india_fcf_dataset.csv")

df.head()
cols = ['Current_price','Fcf 3years','Fcf 5years','Fcf 7years','Fcf 10years','Market_capitalization']

for col in cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')
    df["avg_fcf"] = (
    df["Fcf 3years"] +
    df["Fcf 5years"] +
    df["Fcf 7years"] +
    df["Fcf 10years"]
) / 4

df.head()
growth_rate = 0.10
discount_rate = 0.12
years = 10

df["intrinsic_value"] = df["avg_fcf"] * ((1 + growth_rate) ** years) / ((1 + discount_rate) ** years)
df["undervalued_score"] = df["intrinsic_value"] - df["Current_price"]
df_sorted = df.sort_values(by="undervalued_score", ascending=False)

df_sorted.head(20)
df_sorted.to_csv("undervalued_stocks.csv", index=False)
df_sorted.head(20)
df_sorted.to_csv(r"C:\Users\DELL\Downloads\undervalued_stocks.csv", index=False)
```
